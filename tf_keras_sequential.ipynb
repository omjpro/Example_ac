```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
data = [
    [0.72,0.82,-1],
    [0.91,-0.69,-1],
    [0.03,0.93,-1],
    [0.12,0.25,-1],
    [0.96,0.47,-1],
    [0.8,-0.75,-1],
    [0.46,0.98,-1],
    [0.66,0.24,-1],
    [0.72,-0.15,-1],
    [0.35,0.01,-1],
    [-0.11,0.1,1],
    [0.31,-0.96,1],
    [0.0,-0.26,1],
    [-0.43,-0.65,1],
    [0.57,-0.97,1],
    [-0.72,-0.64,1],
    [-0.25,-0.43,1],
    [-0.12,-0.9,1],
    [-0.58,0.62,1],
    [-0.77,-0.76,1]
]
```

```python
mydata = np.array(data)
x_data = mydata[:,:2]
y_data = mydata[:, -1]
```

```python
plt.scatter(x_data[:,0], x_data[:,1], c = y_data)
plt.show()
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=2, activation="sigmoid", input_shape=(2,)))
model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
optimizer = tf.keras.optimizers.SGD(lr=0.1)
```

```python
model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(x_data, y_data, batch_size=2, epochs=2000)
```
